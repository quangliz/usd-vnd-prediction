```python
import pandas as pd
```

```python
df = pd.read_csv('../usdvnd/usdvnd_cleaned_features.csv')
df.set_index('Ngày', inplace=True)
df.index = pd.to_datetime(df.index)
df.head(20)
```

```python
# create target column for regression
df_regression = df.copy()
df_regression['Target'] = df_regression['Lần cuối'].shift(-1)
df_regression = df_regression.dropna(subset=['Target'])
df_regression['Ngày'] = df_regression.index

# create binary target column for classification
df_classification = df.copy()
df_classification['Target'] = (df_classification['Lần cuối'] > df_classification['Lần cuối'].shift(-1)).astype(int)
df_classification['Ngày'] = df_classification.index

# save data
df_regression.to_csv('../usdvnd/processed/cleaned_features_regression.csv', index=False)
print(df_regression.shape)
df_classification.to_csv('../usdvnd/processed/cleaned_features_classification.csv', index=False)
print(df_classification.shape)
```
