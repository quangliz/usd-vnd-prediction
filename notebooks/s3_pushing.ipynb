```python
import boto3
from pathlib import Path

S3_BUCKET_NAME = 'usd-vnd-prediction'
S3_BUCKET_REGION = 'us-east-1'

PROJECT_ROOT = Path.cwd().parent

print(PROJECT_ROOT)
```

```python
PROCESSED_DATA_FOLDER_PATH = PROJECT_ROOT / 'usdvnd' / 'processed'

s3 = boto3.client(
    's3',
    region_name=S3_BUCKET_REGION,
)

def upload_file_to_s3(
    file_path: Path,
    s3_key: str,
    bucket_name: str = S3_BUCKET_NAME,
):
    """
    Upload a file to S3
    """
    if not file_path.exists():
        raise FileNotFoundError(f"File {file_path} not found")
    s3.upload_file(
        Filename = file_path,
        Bucket = bucket_name,
        Key = f'{s3_key}/{file_path.name}',
    )
    print(f"File {s3_key}/{file_path} uploaded to S3")

# upload_file_to_s3(PROCESSED_DATA_FOLDER_PATH / 'cleaned.csv', s3_key='data')
# upload_file_to_s3(PROCESSED_DATA_FOLDER_PATH / 'feature_engineered.csv', s3_key='data')
# upload_file_to_s3(PROCESSED_DATA_FOLDER_PATH / 'cleaned_features_regression.csv', s3_key='data')
```
